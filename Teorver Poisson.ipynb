$\textbf{Предельная теорема Пуассона}$:  Пусть $\xi_n \sim Binom(n,p_n)$ и $np_n  \rightarrow \lambda>0$ при $n \rightarrow \infty$ тогда $\xi_n \sim Poisson(\lambda)$

```python
import os
from scipy.special import factorial
from scipy.special import comb
import matplotlib
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import numpy as np

def CNK(n,k): #подсчёт числа сочетаний из n по k (где k может быть массивом numpy)
    return comb(n,k,exact=False, repetition=False)

def BINOM(n,k,p):
    return CNK(n,k)*p**k*(1-p)**(n-k)
    
def FACT(k):  #делаем массив факториалов от массива numpy
    k1=np.array(k)*0.0+1
    for i in range(1,np.max(k)+1):
        for j in range(0,np.size(k)):
            if (k[j]>=i):
                k1[j]*=i
    return k1
    
def POISSON(lamb,k): #распределение пуассона l^ke^-l/k!
    #print k, "--"
    #print "fact",FACT(k)
    #print "exp", np.math.exp(-lamb)
    #print "power",np.power(lamb,k)
    return np.power(lamb,k)*np.math.exp(-lamb)/FACT(k)
    
```

```python
count = 30 #число итераций 
maxlamb = 4 #делаем от лямбда=1 до этого включительно (желательно >3)
err = np.zeros((maxlamb+1,count))
for lamb in range(1, maxlamb+1):
    minim = int(lamb // 1+1)
    c=-1
    for n in range(minim, minim+count):
        c+=1
        if lamb>n:
            print "ERROR ERROR ERROR" # противоречие условию
        else:
            pn= lamb*1.0/n

        k = np.array(range(1,min(n, 14)))
        name= 'Binomial('+str(n)+', ' +str(pn)+') and Poisson('+str(lamb)+')'
        plt.title(name)
        plt.xlabel(u'k')
        plt.ylabel(u'P(xi=k)')#, color='green')
        plt.grid(True, color='black')

        plt.bar(k,BINOM(n,k,pn), label = u'Binom')
        plt.plot(k,POISSON(lamb,k), 'R',label = u'Poisson')
        #print "binom  ", BINOM(n,k,pn), type(BINOM(n,k,pn))
        #print "poisson",POISSON(lamb,k),type(POISSON(lamb,k))
        err[lamb-1][c] = np.linalg.norm(POISSON(lamb,k)-BINOM(n,k,pn))
        print u'Невязка =', err[lamb-1][c]
        plt.legend()
        fig = plt #.show()  #если нужно выводить сюда, а не сохранять
        plt.savefig('./poisson/lamb='+str(lamb)+'/n='+str(n)+'.png') #вид: /poisson/lamb=3/n=5.png
        plt.show()
```

```python
# считаем невязку результатов
fig2 = plt.figure()   
grid1 = plt.grid(True)  
plt.title(u'График невязки Биномиального и Пуассоновского от номера итерации (числа n+ \lambda)')
plt.plot(np.linspace(0,count, count), err[0], 'r', label = u'Lamb=1')
plt.plot(np.linspace(0,count, count), err[1], 'g', label = u'Lamb=2')
plt.plot(np.linspace(0,count, count), err[2], 'b', label = u'Lamb=3')
plt.plot(np.linspace(0,count, count), err[3], 'y', label = u'Lamb=4')
plt.legend()
plt.show()
print u'Последнее значение невязки для lamb=1 равно', err[0][count-1]
print u'Последнее значение невязки для lamb=2 равно', err[1][count-1]
print u'Последнее значение невязки для lamb=3 равно', err[2][count-1]
print u'Последнее значение невязки для lamb=4 равно', err[3][count-1]
```
